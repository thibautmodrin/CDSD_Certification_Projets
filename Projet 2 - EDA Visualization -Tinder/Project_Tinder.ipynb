# Elimination des wave 6 à 9 pour les données concernées car systeme de notation incompatibles avec celui des autres waves . je check valeurs aberrantes et NA au fur et à mesure avec cette fonction pour plus de simplicité

```python
def check_features(list_of_features):
  return display(
      'COUNT :',dataset.groupby('wave').count().loc[:,list_of_features].T,
      'MAX :',dataset.groupby('wave').max(numeric_only=True).loc[:,list_of_features].T,
      'MIN :',dataset.groupby('wave').min(numeric_only=True).loc[:,list_of_features].T,
      'NULL:',dataset[list_of_features].isnull().sum().reset_index().T
      )

```

# What are the least desirable attributes in a male partner? Does this differ for female partners?

```python
check_features(['gender','attr1_1','sinc1_1','intel1_1','fun1_1','amb1_1','shar1_1'])
```

```python
dataset_wave=dataset.loc[(dataset['wave']<5) | (dataset['wave']>9) ,:].reset_index(drop=True)
```

```python
df_least=dataset_wave.dropna(subset=['attr1_1','sinc1_1','intel1_1','fun1_1','amb1_1','shar1_1']).groupby('gender').mean(numeric_only=True).loc[:,'attr1_1':'shar1_1']
df_least.index=df_least.index.map({0:'Female',1:'Male'})
df_least.columns=df_least.columns.map({'attr1_1':'Attractiveness','sinc1_1':'Sincerity','intel1_1':'Intelligence','fun1_1':'Fun','amb1_1':'Ambition','shar1_1':'SharedInterest'})
df_least
```

```python
df_least2=dataset[['gender','shar1_1','sinc1_1','intel1_1','fun1_1','amb1_1','attr1_1']].dropna()
df_least2['gender']=df_least2['gender'].map({0:'Female',1:'Male'})
df_least2.columns=df_least2.columns.map({'gender':'gender','attr1_1':'Attractiveness','sinc1_1':'Sincerity','intel1_1':'Intelligence','fun1_1':'Fun','amb1_1':'Ambition','shar1_1':'SharedInterest'})
df_least2.head()
```

```python
fig1=px.pie(df_least,values=df_least.iloc[0],names=df_least.columns,title="Fig 1 : Women's preferences by attributes",width=600,color_discrete_sequence = px.colors.qualitative.Pastel)
fig1.update_layout(
    title_x= 0.5,
    legend_title='Attributes :')
fig1.show()
```

```python
fig2=px.pie(df_least,values=df_least.loc['Male'],names=df_least.columns,title="Fig 2 : Men's preferences by attributes",width=600,color_discrete_sequence = px.colors.qualitative.Pastel)
fig2.update_layout(
    title_x= 0.5,
    legend_title='Attributes :')
fig2.show()
```

```python
fig3=px.box(df_least2,x=['SharedInterest','Sincerity','Intelligence','Fun','Ambition','Attractiveness'],color='gender',title='Fig 3 : Distribution of Preferences by gender',width=900,color_discrete_sequence = px.colors.qualitative.Pastel)
fig3.update_layout(
    title_x= 0.5,
    legend_title='Gender :',
    xaxis_title='Rating',
    yaxis_title=None)
fig3.show()
```

**Interpretation**

Figures 1 show us that the attributes women look for least in men are ambition and shared interests. Figure 2 shows us that ambition and shared interests are also the least desirable attributes in men for women. Moreover, figure 3 shows us that men's disinterest in these two attributes is even more pronounced. We also note that the distribution for all attributes is much more polarized for men towards a strong demand for women's attractiveness.

# How important do people think attractiveness is in potential mate selection vs. its real impact?

```python
check_features(['gender','shar1_1','sinc1_1','intel1_1','fun1_1','amb1_1','attr1_1','dec','attr','sinc','intel','fun','amb','shar','attr7_2','sinc7_2','intel7_2','fun7_2','amb7_2','shar7_2'])
```

```python
df_selected=pd.concat([dataset_wave.loc[dataset_wave['wave']>14,'attr1_1':'shar1_1'],dataset_wave.loc[dataset_wave['wave']>14,'attr7_2':'shar7_2']],axis=1).dropna().mean()
df1=df_selected['attr1_1':'shar1_1']
df2=df_selected['attr7_2':'shar7_2']
df1.name='Before'
df2.name='Review After'
df1.index=df1.index.map({'attr1_1':'Attractiveness','sinc1_1':'Sincerity','intel1_1':'Intelligence','fun1_1':'Fun','amb1_1':'Ambition','shar1_1':'SharedInterest'})
df2.index=df2.index.map({'attr7_2':'Attractiveness','sinc7_2':'Sincerity','intel7_2':'Intelligence','fun7_2':'Fun','amb7_2':'Ambition','shar7_2':'SharedInterest'})
df_selected=pd.concat([df1,df2],axis=1)
df_selected
```

```python
df_selected1=pd.concat([dataset_wave.loc[dataset_wave['wave']>14,'attr1_1':'shar1_1'],dataset_wave.loc[dataset_wave['wave']>14,'attr7_2':'shar7_2']],axis=1).dropna()
df11=df_selected1.loc[:,'attr1_1':'shar1_1']
df21=df_selected1.loc[:,'attr7_2':'shar7_2']
df11['When']='Projected'
df21['When']='Revised'
df11.head()
df11.columns=df11.columns.map({'attr1_1':'Attractiveness','sinc1_1':'Sincerity','intel1_1':'Intelligence','fun1_1':'Fun','amb1_1':'Ambition','shar1_1':'SharedInterest','When':'When'})
df21.columns=df21.columns.map({'attr7_2':'Attractiveness','sinc7_2':'Sincerity','intel7_2':'Intelligence','fun7_2':'Fun','amb7_2':'Ambition','shar7_2':'SharedInterest','When':'When'})
df_selected1=pd.concat([df11,df21],axis=0)
df_selected1.head()
```

```python
df_real=dataset_wave.dropna(subset=['dec','attr','sinc','intel','fun','amb','shar']).groupby('dec').mean(numeric_only=True).loc[:,'attr':'shar'].round(2).T
df_real.columns=df_real.columns.map({0:'No',1:'Yes'})
df_real.columns.name='Decision'
df_real.index=df_real.index.map({'attr':'Attractiveness','sinc':'Sincerity','intel':'Intelligence','fun':'Fun','amb':'Ambition','shar':'SharedInterest'})

df_real
```

```python
df_diff=df_real.loc[:,'Yes']-df_real.loc[:,'No']
df_diff.name='Difference'
df_diff
```

```python
fig4=px.bar(df_real,barmode='group',range_y=[0,10],width=800,title='Fig 4 : Influence of Preferences Partner on Decision',color_discrete_sequence = px.colors.qualitative.Pastel1)

fig4.update_layout(
    xaxis_title='Attributes',
    yaxis_title='Average Rating',
        legend_bordercolor="lightgrey",
    legend_borderwidth=1,
    legend_x=0.88,
    legend_y=1,
    title_x= 0.5)
fig4.add_trace(go.Scatter(x=df_diff.index,y=df_diff,name='Diff',marker_size=15,marker_symbol='square',marker_color="green"))
fig4.show()
```

```python
fig5=px.box(df_selected1,range_y=[0,100],width=700,title='Fig 5 : Projected vs Revised Attributes',color='When',color_discrete_sequence = px.colors.qualitative.Pastel1)
fig5.update_layout(
    xaxis_title='Attributes',
    yaxis_title='Rating',
    title_x= 0.5,
    legend_bordercolor="lightgrey",
    legend_borderwidth=1,
    legend_x=0.8,
    legend_y=0.9,
    legend_title='Look For:')
fig5.show()
```

**Interpretation**

Figure 5 shows us that attractiveness is the most desired attribute among partners before speed dating, but it's also the most widely dispersed, showing heterogeneous opinions on it.
It is also the criterion that receives the most additional points after the date, showing that it is the most under-rated criterion.


Moreover, figure 4 shows us that participants tend to give higher ratings to attractiveness, fun and shared interests for the partner with whom they have decided to continue after the date. This confirms that actractiveness is a key criterion in the possibility of two partners matching and continuing to a second speed dating.

## Are shared interests more important than a shared racial background?

```python
check_features(['match','int_corr','imprace','samerace'])
```

```python
df_share1=dataset.dropna(subset=['match','int_corr','imprace','samerace']).groupby('match')[['int_corr','imprace']].mean().round(2)
df_share1.index=df_share1.index.map({0:'No',1:'Yes'})
df_share1.columns=df_share1.columns.map({'int_corr':'Correlation Interest','samerace':'Same Race','imprace':'Same race Importance'})
df_share1=df_share1.T
df_share1
```

```python
df_share11=dataset.dropna(subset=['match','int_corr','imprace']).loc[:,['int_corr','imprace','match']]
df_share11['match']=df_share11['match'].map({0:'No',1:'Yes'})
df_share11.head()
```

```python

df_share2=dataset.dropna(subset=['match','int_corr','imprace']).groupby('match')['samerace'].value_counts(normalize=True).reset_index(name='Proportion').set_index('match')
df_share2.index=df_share2.index.map({0:'No',1:'Yes'})
df_share2.loc[:,'samerace']=df_share2.loc[:,'samerace'].map({0:'No',1:'Yes'})
df_share2.loc[:,'Proportion']=df_share2.loc[:,'Proportion'].map(lambda x: round(x*100,0))
df_share2
```

```python
fig6 = make_subplots(rows=1, cols=2,subplot_titles=('Correlation Interest','Same Race Importance'))

fig61=px.box(df_share11,x='match',y='int_corr', range_y=[0,1],width=500,color='match',color_discrete_sequence = px.colors.qualitative.Set2)

fig62=px.box(df_share11,range_y=[0,10],x='match',y='imprace',width=500,color='match',color_discrete_sequence = px.colors.qualitative.Set2)

for trace in fig61.data:
    fig6.add_trace(trace, row=1, col=1)

for trace in fig62.data:
    fig6.add_trace(trace, row=1, col=2)

fig6.update_layout(
    height=600,
    width=1000,
    title_text="Fig 6 : Disribution of Interests Correlation and Race Importance by Match ",
    title_x=0.5,
    xaxis_title="Match",
    xaxis2_title="Match",
    yaxis_title="Coeff. Corr.",
    yaxis2_title="Rating",
    legend_title="Match",
    legend_visible=False


    )
fig6.show()
```

```python
fig7=px.bar(df_share2,y='Proportion',color='samerace', title='Fig 7 : Race Share Distribution by Match',color_discrete_sequence = px.colors.qualitative.Set2)
fig7.update_layout(
    title_x= 0.5,
    width=500,
    legend_title= 'Get Same Race :',
    yaxis_title='Distribution',
    xaxis_title='Match',
    yaxis_ticksuffix='%'
    )
fig7.show()
```

**Interpretation**

Figure 6 shows us that the correlation of common interests between partners does not significantly impact the possibility of a match.
We also see that the importance of having the same race does not significantly impact the match, however we see that people who have given a rating between 5 and 6 may see their chance of entering a match reduced.
Figure 7 also confirms that the proportion of participants having the same race or not is identical whether the participants have matched or not. which confirms to us that having the same race is not an essential criterion in the possibility of matching.

In conclusion, whether participants share the same race or the same interests does not significantly affect their chances of matching.

# Can people accurately predict their own perceived value in the dating market?

```python
check_features(['attr_o','sinc_o','intel_o','fun_o','amb_o','attr3_1','sinc3_1','intel3_1','fun3_1','amb3_1'])
```

```python
name_index=['Attractive','Sincere','Intelligent','Fun','Ambitious']
name_col=['Estimate','Real','Diff']

do=dataset.dropna(subset=['attr_o','sinc_o','intel_o','fun_o','amb_o']).loc[:,'attr_o':'amb_o'].median()
do.index=name_index
d1=dataset.dropna(subset=['attr3_1','sinc3_1','intel3_1','fun3_1','amb3_1']).loc[:,'attr3_1':'amb3_1'].median()

d1.index=name_index

df_predict=pd.concat([d1,do,(d1-do)],axis=1)
df_predict.columns=name_col
df_predict
```

```python
fig8=px.bar(df_predict.iloc[:,[0,1]],barmode='group',range_y=[-10,10],width=900,title='Fig 8 : Ratings of partner after Dating vs Own Estimate',color_discrete_sequence = px.colors.qualitative.Plotly)

fig8.update_layout(
    xaxis_title=None,
    yaxis_title='Average Rating',
    legend_title=None,
    title_x= 0.5)
fig8.add_trace(go.Scatter(x=df_predict.index,y=df_predict.iloc[:,2],name='Over vs Under Est.',marker_size=15,marker_symbol='square',marker_color="green"))
fig8.show()
```

**Interpretation**

If we compare the ratings given by a participant's partner with the ratings they give themselves for all preferences. Figure 8 shows us that, on average across all attributes, participants over-rate by 1 point their own perceived value in the dating market

# In terms of getting a second date, is it better to be someone's first speed date of the night or their last?

```python
check_features(['order','round','match','date_3','numdat_3','num_in_3'])
```

```python
data_order=dataset[['order','match','date_3']].dropna(subset=['order','match','date_3'])
```

```python
df_first=data_order.groupby(['order'])['match'].mean().reset_index(drop=False).round(2)
df_first.columns=['Order','Match']
df_first1=data_order.groupby(['order'])['date_3'].mean().reset_index(drop=False).round(2)
df_first1.columns=['Order','Second_Date_Over_All']
```

```python
fig9=px.scatter(df_first,x='Order',y='Match',range_y=[0,1],width=900,title='Fig 9 : Average match by Order of the Date during the Night',trendline='ols',trendline_color_override='red')
fig9.update_layout(
    xaxis_title='Order of the date',
    yaxis_title='Average Match',
    title_x= 0.5)
fig9.show()
```

```python
fig10=px.scatter(df_first1,x='Order',y='Second_Date_Over_All',range_y=[0,1],width=900,title='Fig 10 : Average Second Date Over All by Order of the Date during the Night',trendline='ols',trendline_color_override='red')
fig10.update_layout(
    xaxis_title='Order of the date',
    yaxis_title='Average Second Date Over All',
    title_x= 0.5)
fig10.show()
```

```python
data_order1=dataset[['order','match','date_3','round']].dropna(subset=['order','match','date_3']).sort_values(by='round')
data_order1['round']=data_order1['round'].map(lambda x : str(x))
```

```python
fig11=px.scatter(data_order1,x='order',y='match',color='round',range_y=[0,1],width=900,title='Fig 11 : Average Match by Order during the night and by number of round',trendline='ols',trendline_color_override='red')
fig11.update_layout(
    xaxis_title='Order of the date',
    yaxis_title='Average Match',
    legend_title='Number of round',
    title_x= 0.5)
fig11.update_traces(marker=dict(color='blue'))
fig11.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')
fig11.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgray')

fig11.update_traces(marker=dict(opacity=0))


fig11.show()
```

**Interpretation**

Figure 9 shows that the average number of matches is not significantly affected by the order of the date during the evening. However, figure 11 also shows us that the more rounds there are in the evening and the later the date, the average number of matches between participants tends to decrease slightly.

Conversely, if we take as a reference the number of second dates that took place on all waves for a participant, figure 10 shows us that the later the date, the higher the average number of second dates is. However, this is biased, since the second date did not necessarily take place at the end of the event.

If we refer to this last graph, we can say : It's better to be someone's last speed date of the night due to get a second date.
